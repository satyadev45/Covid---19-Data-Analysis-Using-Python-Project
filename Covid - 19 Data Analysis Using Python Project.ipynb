```python
covid.columns = [c.strip() for c in covid.columns]

num_cols = ['TotalCases','TotalDeaths','TotalRecovered','ActiveCases','TotalTests']
for col in num_cols:
    covid[col] = pd.to_numeric(covid[col], errors='coerce')

covid.dropna(subset=['TotalCases'], inplace=True)
covid = covid[covid['TotalCases'] > 0]
```

```python
total_cases = covid['TotalCases'].sum()
total_deaths = covid['TotalDeaths'].sum()
total_recovered = covid['TotalRecovered'].sum()

death_rate = (total_deaths / total_cases) * 100
recovery_rate = (total_recovered / total_cases) * 100

print("ðŸ“Œ Global KPIs")
print("Total Cases:", total_cases)
print("Total Deaths:", total_deaths)
print("Total Recovered:", total_recovered)
print("Death Rate: %.2f%%" % death_rate)
print("Recovery Rate: %.2f%%" % recovery_rate)
```

# Top 15 Countries Total Cases:

```python
top15 = covid.sort_values(by='TotalCases', ascending=False).head(15)

plt.figure(figsize=(18,6))
plt.bar(top15['Country/Region'], top15['TotalCases'], color='royalblue')
plt.xticks(rotation=45, ha='right')
plt.title("Top 15 Countries - Total COVID-19 Cases")
plt.ylabel("Total Cases")
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()

plt.show()
```

# Top 15 Countries â€“ Deaths:

```python
top15_deaths = covid.sort_values(by='TotalDeaths', ascending=False).head(15)

plt.figure(figsize=(18,6))
sns.barplot(x=top15_deaths['Country/Region'], y=top15_deaths['TotalDeaths'], palette="Reds_r")
plt.xticks(rotation=45, ha='right')
plt.title("Top 15 Countries - Total COVID-19 Deaths")
plt.ylabel("Total Deaths")
plt.tight_layout()
plt.show()
```

# Bubble Chart Cases by Continent:

```python
plt.figure(figsize=(18,8))

sizes = (top15['TotalCases'] / top15['TotalCases'].max()) * 3000
colors = sns.color_palette("viridis", len(top15))

plt.scatter(top15['Continent'], top15['TotalCases'],
            s=sizes, c=colors, alpha=0.7)

plt.xlabel("Continent")
plt.ylabel("Total Cases")
plt.title("COVID-19 Bubble Plot â€” Top 15 Countries")
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

# Heatmap (Professional Analysis):

```python
# plt.figure(figsize=(10,6))
sns.heatmap(covid[num_cols].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of COVID-19 Indicators")
plt.tight_layout()
plt.show()
```

# Continent-wise Total Cases (Pie Chart):

```python
# Group continent-wise totals
continent_cases = covid.groupby("Continent")["TotalCases"].sum()

plt.figure(figsize=(10, 8))
plt.pie(continent_cases, labels=continent_cases.index, autopct="%1.1f%%",
        startangle=140, shadow = True, wedgeprops={"edgecolor": "black"})

plt.title("COVID-19 Total Cases Distribution by Continent")
plt.tight_layout()
plt.show()
```

# Cases vs Deaths Scatter (Risk Map):

```python
plt.figure(figsize=(14, 8))

sns.scatterplot(
    data=covid,
    x="TotalCases",
    y="TotalDeaths",
    hue="Continent",
    size="TotalCases",
    sizes=(20, 800),
    alpha=0.7,
)

plt.title("Cases vs Deaths â€” COVID-19 Risk Scatter Map")
plt.xlabel("Total Cases")
plt.ylabel("Total Deaths")
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
